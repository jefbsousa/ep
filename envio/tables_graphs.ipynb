```python
import pandas as pd
import matplotlib
```

```python
df = pd.read_csv("results_1.csv", sep=";", header=None, 
                 names=['Método', 'N', 'núm. iterações', 'tempo(seg)', 'erro máx'])
```

```python
df_1A_10e5 = df.loc[1:28]
df_1B_10e5 = df.loc[30:57]

df_1A_10e8 = df.loc[59:86]
df_1B_10e8 = df.loc[88:115]
```

<h4>1A com 10-5 , tempo de execução</h4>

```python
df_1A_10e5[['Método', 'N', 'tempo(seg)']].pivot(index='N', columns='Método', values='tempo(seg)').plot()
```

<h4>1A com 10-5 , erro máximo</h4>

```python
df_1A_10e5[['Método', 'N', 'erro máx']].pivot(index='N', columns='Método', values='erro máx').plot()
```

<h4>1B com 10-5 , tempo de execução</h4>

```python
df_1B_10e5[['Método', 'N', 'tempo(seg)']].pivot(index='N', columns='Método', values='tempo(seg)').plot()
```

<h4>1B com 10-5 , e erro máximo</h4>

```python
df_1B_10e5[['Método', 'N', 'erro máx']].pivot(index='N', columns='Método', values='erro máx').plot()
```
