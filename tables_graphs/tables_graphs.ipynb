```python
import pandas as pd
import matplotlib
```

```python
df = pd.read_csv("../C/results_1.csv", sep=";", header=None, 
                 names=['Método', 'N', 'núm. iterações', 'tempo(seg)', 'erro máx'])
```

```python
df_1A_10e5 = df.loc[1:21]
df_1B_10e5 = df.loc[22:42]
```
