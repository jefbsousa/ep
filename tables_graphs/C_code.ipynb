```python
import pandas as pd
import matplotlib
```

```python
df = pd.read_csv("../C/results_1A.csv", sep=";", header=None, 
                 names=['Método', 'N', 'núm. iterações', 'tempo(seg)', 'erro máx'])
```

```python
df.head(4)
```

```python
df_pivot = df[['Método', 'N', 'tempo(seg)']].pivot(index='N', columns='Método', values='tempo(seg)')

df_pivot.plot()
```

```python
df_p = df[['Método', 'N', 'erro máx']].pivot(index='N', columns='Método', values='erro máx')

df_p.plot()
```
